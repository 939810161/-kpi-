```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## 加载数据的函数

```python
def loaddata():
    data = np.loadtxt('data/data1.txt',delimiter=',')
    n = data.shape[1]-1 #特征数
    X = data[:,0:n]
    y = data[:,-1].reshape(-1,1)
    return X,y
```

```python
X,y = loaddata()
```

## 1、线性回归(最小二乘实现)
linear_model中的LinearRegression()方法可以实现线性回归

```python
model1 = linear_model.LinearRegression()
```

fit()方法可以加载训练数据并进行训练

```python
#调用fit方法
model1.fit(X,y)
```

intercept_属性可以输出$\theta_0$的值，coef_属性可以输出$\theta_1$到$\theta_n$的值

```python
#输出系数
print(model1.coef_ )
#输出截距
print(model1.intercept_)
```

```python
plt.scatter(X,y)
y_hat = model1.predict(X)
plt.plot(X,y_hat)
plt.show()
```

## 2、Ridge
linear_model中的Ridge方法可以实现Ridge回归

注意：这里的alpha表示正则化强度

normalize设置为True表示对训练数据进行标准化

```python
model2 = linear_model.Ridge(alpha=0.01, normalize=True)
model2.fit(X,y)
#输出系数
print(model2.coef_ )
#输出截距
print(model2.intercept_)
plt.scatter(X,y)
y_hat = model2.predict(X)
plt.plot(X,y_hat)
plt.show()
```

## 3、LASSO
linear_model中Lasso()方法可以实现Lasso回归

注意：这里的alpha表示正则化强度

normalize设置为True表示对训练数据进行标准化

```python
model3 = linear_model.Lasso(alpha=0.01, normalize=True)
model3.fit(X,y)
#输出系数
print(model3.coef_ )
#输出截距
print(model3.intercept_)
plt.scatter(X,y)
y_hat = model3.predict(X)
plt.plot(X,y_hat)
plt.show()
```

## 4、ElactisNet
linear_model中的ElasticNet()方法可以实现弹性网

注意：这里的alpha表示正则化强度

normalize设置为True表示对训练数据进行标准化

```python
model4 = linear_model.ElasticNet(alpha=0.01, normalize=True)
model4.fit(X,y)
#输出系数
print(model4.coef_ )
#输出截距
print(model4.intercept_)
plt.scatter(X,y)
y_hat = model4.predict(X)
plt.plot(X,y_hat)
plt.show()
```
